### Bagian batang tubuh

```python
# Fungsi untuk menghilangkan bagian yang tidak diinginkan setelah parsing
def text_preprocessing_1(input):
  temp = re.sub('\\nBAB.*?\s\.\.\.\\n-\d+-', '', input)
  temp = re.sub('\\nBagian.*?\s\.\.\.\\n-\d+-', '', temp)
  temp = re.sub('\\nParagraf.*?\s\.\.\.\\n-\d+-', '', temp)
  temp = re.sub('\\nPasal.*?\s\.\.\.\\n-\d+-', '', temp)
  temp = re.sub('\\n\(\d+\).*?…\\n-\d+-', '', temp)
  temp = re.sub('\\n\d+.*?\.\.\.\\n-\d+-', '', temp)
  temp = re.sub('\\n[a-z].*?…\\n-\d+-', '', temp)
  output = re.sub('\\n-\d+-', '', temp)
  return output
```

```python
# Fungsi untuk 'membersihkan' teks
def text_preprocessing_2(input):
  text = re.sub(r'\n', ' ', input) # mengganti jarak antar baris menjadi sebuah spasi
  text = re.sub(r'BAB.*', '', text) # menghilangkan bagian "BAB ..." yang terdeteksi saat parsing
  text = re.sub(r'Bagian.*', '', text) # menghilangkan bagian "Bagian ..." yang terdeteksi saat parsing
  text = re.sub(r'Paragraf.*', '', text) # menghilangkan bagian "Paragraf ..." yang terdeteksi saat parsing

  # menghilangkan jarak antar baris yang berlebih
  text = re.sub(' +', ' ', text.replace('\n', ' '))
  output = text.strip()
  return output
```

```python
def article_text_processing(pasal, judul_peraturan):
  # Menghilangkan bagian yang tidak diinginkan setelah parsing
  text = re.sub(r'#\s\[BAB.*', '', pasal)
  text = re.sub(r'##\s\[Bagian.*', '', text)
  text = re.sub(r'##\s\[Paragraf.*', '', text)
  text = re.sub(r'\(http[s]?://\S+', '', text)
  text = re.sub(r'[*#]', '', text)
  text = re.sub(r'\n', ' ', text)

  # Memodifikasi teks (menambahkan identifier tertentu)
  # Harapannya suatu LLM dapat diinstruksikan untuk mengenali isi dari tiap nomor pasal berdasarkan identifier tersebut
  temp = re.findall(r'\[Pasal \d*\]', text)[0]
  pasal_id = re.sub(r'[\[\]]', '', temp)
  text = re.sub(r'\[Pasal \d*\]', '', text)
  identifier = "[Berikut adalah isi {} pada {}]".format(pasal_id, judul_peraturan)
  text = identifier + " " + text

  # Menghilangkan line break yang terlalu banyak
  text = re.sub(' +', ' ', text.replace('\n', ' '))
  text = text.strip()
  return text
```

> Bisa dilihat pada contoh di atas, bahwa terdapat bagian yang tidak diinginkan (unwanted parts) setelah proses parsing. Penanganannya akan dilakukan secara manual.

```python
'''
daftar entri yang mengandung unwanted parts:
(\nBAB II  \nKEPESERTAAN...)
- list_pasal[3]
- list_pasal[10]
- list_pasal[17]
- list_pasal[41]
- list_pasal[43]
- list_pasal[44]
- list_pasal[45]
- list_pasal[47]

(\nBagian Kedua  \nTata Cara...)
- list_pasal[4]
- list_pasal[13]
- list_pasal[20]
- list_pasal[24]
- list_pasal[29]
- list_pasal[34]
'''
```

```python
indices_1 = [3, 10, 17, 41, 43, 44, 45, 47]
indices_2 = [4, 13, 20, 24, 29, 34]
```

```python
for i in range(len(list_pasal)):
  if i in indices_1:
    temp = list_pasal[i].partition("  \n \nBAB")
    list_pasal[i] = temp[0]
  elif i in indices_2:
    temp = list_pasal[i].partition("  \n \nBagian Kedua")
    list_pasal[i] = temp[0]
```

```python
list_pasal[3]
```

```python
list_pasal_modified = []
for i in range(2, len(list_pasal)):
  # remove multiple \n and whitespace
  temp = re.sub(' +', ' ', list_pasal[i].replace('\n', ' ')).strip()
  # modify the text so that (hopefully) the LLM can be instructed to easily identify the article number
  list_pasal_modified.append("[Berikut adalah isi Pasal " + str(i) + " pada " + "PP 37/2021" + "] " + temp)
```

```python
# List yang mengandung seluruh potongan peraturan
PP_37_2021 = pasal_1_modified + list_pasal_modified
len(PP_37_2021)
```
